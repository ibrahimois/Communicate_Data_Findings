```python
#load the data using pandas
go_bike_df = pd.read_csv("201902-fordgobike-tripdata.csv")
go_bike_df.head()
```

```python
#a brief info about the dataframe
go_bike_df.info()
```

As we can see start_time and end_time are from type object which will not grant us much use, we will simply convert them to datetime64 for more usefull tools later on 

also we will convert the duration_sec into new column duratuin_hour, to be easily readable

```python
#convert the dtype of the columns start_time & end_time to => datetime64[ns]
go_bike_df["start_time"] = go_bike_df["start_time"].astype("datetime64[ns]")
go_bike_df["end_time"] = go_bike_df["end_time"].astype("datetime64[ns]")
```

```python
#create a new column that will help us in the analysis process
go_bike_df["duration_hour"] = go_bike_df["duration_sec"] / 3600 # Dividing by 3600 to convert seconds to hours
go_bike_df["duration_hour"]
```

```python
go_bike_df.info()
```

```python
go_bike_df.describe()
```

```python
go_bike_df["start_time"].dt.hour
```

```python
go_bike_df["end_time"].dt.hour
```

we can see that the min member_birth_year is 1878 which can not be possable so we can do some snippin to clean the data for more accurate results

we can also create a new column called member_age to easily identify the age of the biker

```python
#clear the dataframe by dropping rows that have member_birth_year lower that 1940
go_bike_df = go_bike_df[go_bike_df["member_birth_year"] > 1940]
```

```python
#create a new column that will help us in the analysis process
go_bike_df["member_age"] = 2021 - go_bike_df["member_birth_year"] # 2021 minus the member birth year will give us thier approximate age
go_bike_df["member_age"]
```

## Univariate Exploration

> I will start investigating the main features that i mentioned in the above article.

```python
#a plot for the number of members from a given age 
go_bike_df[['member_age']].plot(kind='hist', figsize=(10,5), grid=True)
plt.title("Number of members from a given age")
plt.xticks(np.arange(20)*5)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

by looking at the above plot we can see that the majority of our members are between 45-20, also our members peek between 30-35

```python
#a plot for the number of users from a given type
plt.figure(figsize=(10,5))
go_bike_df['user_type'].hist(bins=3)
plt.title("number of users from a given type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

by looking at the above histogram we can see that 80% of our users are members

```python
#a plot for number of users from a given gender
plt.figure(figsize=(10,5))
go_bike_df['member_gender'].hist(bins=5)
plt.title("number of users from a given gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

by lookin at the above histogram we can see that 80% from our members are men 

```python
#a plot for the starting hour for each trip
plt.figure(figsize=(10,5))
go_bike_df["start_time"].dt.hour.hist()
plt.title("Starting hour for trips")
plt.xticks(np.arange(24))
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()
```

by looking at the above histogram we can see that the majority of our trips start at 7AM and 5PM 

```python
def plotinfo(title, ticks, xlabel, ylabel):
    plt.title(title)
    plt.xticks(ticks)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
#a plot for the ending hour of each trip
plt.figure(figsize=(10,5))
go_bike_df["end_time"].dt.hour.hist()
plotinfo("Ending hour for trips", np.arange(24), "Hour", "Count")
plt.show()
```

by lookin at the above histogram we can see that the end time is fairly simmilar to the start time, my geuss is that it is this way because most of the trips are about an hour or so, but we can see after plotting the duration of our trips

```python
#a plot for each trip for a given day
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sb.countplot(x = go_bike_df.start_time.dt.day_name(), data = go_bike_df, order = day_order, color=sb.color_palette()[0])
plt.xlabel("Day")
plt.title("Number of trips in each day of the week")
```

we can see that week-ends have lower trips

```python
#a plot for duration per trip in hours
bins = np.arange(0, 1.5, 0.02)
ticks = np.arange(0, 1.5, 0.1)
plt.figure(figsize=(10,5))
go_bike_df["duration_hour"].hist(bins=bins)
plt.title("duration per trip in hours")
plt.xticks(ticks)
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()
```

looking at this graph it explains why the end_time and the start_time looked similur, majority of trips were between 0.01h to 0.25h

## Bivariate Exploration

> i will start migrating some features togther to see the relation between them and plot them

```python
#a plot for duration in hours per age 
ticks = np.arange(0, 2.5, 0.1)
plt.figure(figsize=(10,5))
plt.ylim(ymin=0)
plt.yticks(ticks)
plt.scatter(go_bike_df["member_age"], go_bike_df["duration_hour"], alpha=0.25, marker=".")
plt.title("Duration in hours per age")
plt.xlabel("Age")
plt.ylabel("Hour")
plt.show()
```

Here we can see that membere from age 20-60 take trips in duration for around 0.1-1 Hour 

```python
#a plot for duration in hours for each day in the week
day = go_bike_df["start_time"].dt.day_name()
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
ticks = np.arange(0, 2.5, 0.1)
plt.figure(figsize=(10,5))
plt.ylim(ymin=0)
plt.yticks(ticks)
sb.boxenplot(x=go_bike_df.start_time.dt.day_name(),y=go_bike_df.duration_hour, order=day_order)
plt.title("Duration in hours for each day of the week")
plt.xlabel("Day")
plt.ylabel("Hour")
plt.show()
```

even though the week-ends effect the amount of trips (as we saw in previuos graphs) the duration is fairly the same

## Multivariate Exploration

> i will look for some relation betweet day duration and gender, also i will look for some relation between age duration and gender 

```python
#a plot to show the relation between duration in hours for each day of the week for each gender
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
plt.figure(figsize=(12,8))
sb.barplot(x=go_bike_df.start_time.dt.day_name(), y= go_bike_df.duration_hour, hue=go_bike_df.member_gender, data=go_bike_df, order=day_order)
plt.title("the relation between duration in hours for each day of the week for each gender")
plt.xlabel("Day")
plt.ylabel("Hour")
plt.show()
```

"Other" gender seem to be standing out in this plot, on average it seems like they take more duration than the others

```python
#a plot to show the relation between duration in hours and the type of user in a given day
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

plt.figure(figsize=(12,8))
sb.barplot(x=go_bike_df.start_time.dt.day_name(), y= go_bike_df.duration_hour, hue=go_bike_df.user_type, data=go_bike_df, order=day_order);
plt.title("the relation between duration in hours and the type of user in a given day")
plt.xlabel("Day")
plt.ylabel("Hour")
plt.show()
```

"Customers"  seem to be standing out in this plot, on average it seems like they take more duration than the Subscribers
